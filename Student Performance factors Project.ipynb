## Data Cleaning

```python
# Filling categorical missing values with mode
for col in df.select_dtypes(include='object').columns:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

Mode represents the most frequent category.  
For categorical data, replacing missing values with the mode preserves category distribution.  

```python
# Filling numerical missing values with median
for col in df.select_dtypes(include=np.number).columns:
    df[col].fillna(df[col].median(), inplace=True)
```

Median is preferred because:  

It is less affected by outliers  
Provides stable central tendency  
Preserves data distribution better than mean

## Statistical Testing  

EDA alone shows trends.  
Statistical testing validates whether those trends are statistically significant.  
This makes analysis more reliable and professional.  

```python
from scipy.stats import pearsonr
from scipy.stats import ttest_ind
from scipy.stats import f_oneway
```

### Pearson Correlation Test
Pearson correlation measures:  
Relationship strength   
Statistical significance  

```python
corr, p_value = pearsonr(
    df['Hours_Studied'],
    df['Exam_Score']
)

print(corr)
print(p_value)
```

Understanding P-Value  
p < 0.05 → statistically significant  
p > 0.05 → not statistically significant  

### T-Test

```python
internet_yes = df[
    df['Internet_Access'] == 'Yes'
]['Exam_Score']

internet_no = df[
    df['Internet_Access'] == 'No'
]['Exam_Score']

stat, p = ttest_ind(
    internet_yes,
    internet_no
)

print(p)
```

T-test compares means between two groups.  
This helps determine whether internet access significantly affects exam performance.  

### ANOVA Test

```python
low = df[
    df['Motivation_Level'] == 'Low'
]['Exam_Score']

medium = df[
    df['Motivation_Level'] == 'Medium'
]['Exam_Score']

high = df[
    df['Motivation_Level'] == 'High'
]['Exam_Score']

f_stat, p = f_oneway(
    low,
    medium,
    high
)

print(p)
```

ANOVA compares means across multiple groups.  
This helps determine whether motivation level significantly affects academic performance.
